```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/JBoundless/team-shooting-viz/main/TeamShootingStats%20-%20Sheet1-2.csv')

# Teams
TeamRow1 = df[df.TeamRow1 == 'TeamRow1']
TeamRow2 = df[df.TeamRow1 == 'TeamRow2']
TeamRow3 = df[df.TeamRow1 == 'TeamRow3']
TeamRow4 = df[df.TeamRow1 == 'TeamRow4']
TeamRow5 = df[df.TeamRow1 == 'TeamRow5']

# Distances
LessThan5_FG = df[df.LessThan5_FG == 'LessThan5_FG']
Five_Nine_FG = df[df.Five_Nine_FG == 'Five_Nine_FG']
Ten_Fourteen_FG = df[df.Ten_Fourteen_FG == 'Ten_Fourteen_FG']
Fifteen_Nineteen_FG = df[df.Fifteen_Nineteen_FG == 'Fifteen_Nineteen_FG']
Twenty_TwentyFour_FG = df[df.Twenty_TwentyFour_FG == 'Twenty_TwentyFour_FG']
TwentyFive_TwentyNine_FG = df[df.TwentyFive_TwentyNine_FG == 'TwentyFive_TwentyNine_FG']

plt.bar(df.TeamRow1, df.LessThan5_FG, color="blue")
plt.title("FG% From Less Than 5 Ft")
plt.xticks(rotation = 'vertical')
plt.show()
```
